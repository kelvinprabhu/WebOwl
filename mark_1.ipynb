```python
from PIL import Image
import requests

from transformers import CLIPProcessor, CLIPModel

model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

url = "http://images.cocodataset.org/val2017/000000039769.jpg"
image = Image.open(requests.get(url, stream=True).raw)

inputs = processor(text=["a photo of a cat", "a photo of a dog"], images=image, return_tensors="pt", padding=True)

outputs = model(**inputs)
logits_per_image = outputs.logits_per_image  # this is the image-text similarity score
probs = logits_per_image.softmax(dim=1)  # we can take the softmax to get the label probabilities
```

```python
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import torch
import os

# Use CUDA if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load BLIP model and processor
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base").to(device)

# Directory containing images
image_dir = "web_owl_output"

# Loop through image files and generate captions
for file in os.listdir(image_dir):
    if file.lower().endswith(('.jpg', '.jpeg', '.png', '.webp')):
        image_path = os.path.join(image_dir, file)
        image = Image.open(image_path).convert('RGB')

        inputs = processor(image, return_tensors="pt").to(device)
        with torch.no_grad():
            out = model.generate(**inputs)

        caption = processor.decode(out[0], skip_special_tokens=True)
        print(f"{file}: {caption}")
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration
from langchain.text_splitter import RecursiveCharacterTextSplitter
import torch
import os

# Setup
url = "https://medium.com/user-experience-design-1/cracking-the-code-of-vibe-coding-124b9288e551"
output_dir = "web_owl_output"
os.makedirs(output_dir, exist_ok=True)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# BLIP model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base").to(device)

# Scrape web page

response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Get and save text
text = soup.get_text(strip=True)
with open(os.path.join(output_dir, "page_text.txt"), "w", encoding="utf-8") as f:
    f.write(text)

# Get image URLs
img_tags = soup.find_all("img")
img_urls = [urljoin(url, img.get("src")) for img in img_tags if img.get("src")]

# Caption and index images
image_captions = []
for i, img_url in enumerate(img_urls):
    try:
        img_data = requests.get(img_url).content
        ext = os.path.splitext(img_url)[-1][:5] or ".jpg"
        img_path = os.path.join(output_dir, f"image_{i}{ext}")
        with open(img_path, "wb") as f:
            f.write(img_data)

        image = Image.open(img_path).convert("RGB")
        inputs = processor(image, return_tensors="pt").to(device)
        with torch.no_grad():
            out = model.generate(**inputs)
        caption = processor.decode(out[0], skip_special_tokens=True)

        image_captions.append({
            "filename": f"image_{i}{ext}",
            "caption": caption,
            "path": img_path
        })

    except Exception as e:
        print(f"Failed {img_url}: {e}")

# Split text for RAG
text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=40)
chunks = text_splitter.split_text(text)

# Output sample
print("Chunks:", chunks[:2])
print("Image Captions:", image_captions[:2])
```

```python
import os
import cv2
import yt_dlp

# Step 1: Download video with yt-dlp
video_url = 'https://www.youtube.com/watch?v=dQw4w9WgXcQ'
output_filename = 'video.mp4'

ydl_opts = {
    'format': 'best',
    'outtmpl': output_filename
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([video_url])

print("Video downloaded!")

# Step 2: Extract one frame every 3 seconds
cap = cv2.VideoCapture(output_filename)
fps = cap.get(cv2.CAP_PROP_FPS)
interval = int(fps * 3)

os.makedirs("frames", exist_ok=True)

frame_count = 0
saved = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    if frame_count % interval == 0:
        cv2.imwrite(f"frames/frame_{saved}.jpg", frame)
        saved += 1
    frame_count += 1

cap.release()
print(f"Saved {saved} frames at 3-second intervals.")
```
